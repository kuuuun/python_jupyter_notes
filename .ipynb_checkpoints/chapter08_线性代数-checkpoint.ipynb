# chapter08 线性代数

```python
import numpy as np
```

矩阵 * 矩阵 就是矩阵乘法  
矩阵 * 向量 就是矩阵的变换  
向量 * 向量 就是向量的点积

```python
# 定义两个矩阵
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

print("矩阵 A:\n", A)
print("矩阵 B:\n", B)

# 1. 逐元素乘法 (*)
# 对应位置相乘：1*5, 2*6, ...
print("\n逐元素乘法 (A * B):\n", A * B)

# 2. 矩阵乘法 (@)
# 行乘列：1*5+2*7, 1*6+2*8, ...
print("\n矩阵乘法 (A @ B):\n", A @ B)

# 3. 向量的点积
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
# 1*4 + 2*5 + 3*6 = 32
print(f"\n向量内积 (v1 @ v2): {v1 @ v2}")
```

### 向量 

#### 含义：`[5,3]` 代表 水平方向移动5个单位，竖直方向移动3个单位

- 向量相加，vec_a + vec_b 是 vec_a 的起点到 vec_b 的终点
- 向量相减，vec_a - vec_b 是 vec_b 的终点到 vec_a 的终点, 相当于b点到a点的向量
- 乘标量是缩放

### 向量的长度

又叫向量的模, 

- 向量/向量的模，得到方向相同，长度为一的单位向量。归一化

- `np.linalg.norm()
- 点积法，np.sqrt(v @ v)

```python
vt1=np.array([1,0,0]) # x axis
vt2=np.array([0,1,0]) # y axis
vt3=np.array([1,1,0]) # 45deg axis

print(f"vt2-vt1: {vt2-vt1}")
print(f"vt1+vt2: {vt1+vt2}")
assert (vt3==vt1+vt2).all(), "not equal"
assert (vt3==vt1+vt2).any(), "not equal"
print(f"v1 length: {np.linalg.norm(vt1)}")
print(f"v3 length: {np.linalg.norm(vt3)}")
print(f"v3 length: {np.sqrt(vt3@vt3)}")

# assert (vt3==vt1).all(), "not equal"
```

### 叉积是垂直向量

### 点积是标量
向量点积的几何定义为： $ \vec{a} \cdot \vec{b} = \|\vec{a}\| \|\vec{b}\| \cos\theta $ 
- 向量a，b的模长 乘 cos(angle)
> a@b a是单位向量，才能得出b的投影, 也就是说要得到b的投影，需要除以a的模

```python
vt1=np.array([1,0,0]) # x axis
vt2=np.array([0,1,0]) # y axis
vt3=np.array([1,1,0]) # 45deg axis
print(f"\n向量内积 vt1 @ vt2: {vt1@vt2}")
print(f"\n向量内积 vt3 @ vt1: {vt3@vt1}")
print(f"\n向量内积 vt1 @ vt3: {vt1@vt3}")
print(f"\n向量内积 vt1 在 vt3 上的投影: {vt1@(vt3/np.sqrt(vt3@vt3))}")

print(f"\n向量叉积 vt3, vt1: {np.cross(vt3,vt1)}")
print(f"\n向量叉积 vt3, vt1: {np.cross(vt1,vt3)}")
```
