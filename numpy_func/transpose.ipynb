# transpose

```python
# 2D transpose
import numpy as np
arr = np.arange(6).reshape(2,3)
print(f"origin:\n{arr}")
print(f"transpose:\n{arr.transpose((1,0))}")
print(f"transpose:\n{arr.T}")
```

> axis0=i was 2 blocks, axis1=j was 3blocks, axis2=k was 4blocks

```python
# 3D tanspose
arr = np.arange(24).reshape(2,3,4)
print(f"origin:\n{arr}")
# ignore k, transpose i and j
print(f"transpose:\n{arr.transpose((1,0,2))}")
# shape(2,4,3)
# k轴换为j轴, shape:4->3. 在j轴的三组里，每组挑一个。共8个，再分2大组，每组4个
# shape(3,4,2)
print(f"transpose:\n{arr.transpose((0,2,1))}")
# k轴换为i轴, shape:4->2. 在i轴的两组里，每组挑一个。共12个，在分3大组，每组4个 
print(f"transpose:\n{arr.transpose((1,2,0))}")
```
