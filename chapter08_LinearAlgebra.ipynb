```python
import numpy as np
```

çŸ©é˜µ * çŸ©é˜µ å°±æ˜¯çŸ©é˜µä¹˜æ³•  
çŸ©é˜µ * å‘é‡ å°±æ˜¯çŸ©é˜µçš„å˜æ¢  
å‘é‡ * å‘é‡ å°±æ˜¯å‘é‡çš„ç‚¹ç§¯

```python
# å®šä¹‰ä¸¤ä¸ªçŸ©é˜µ
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

print("çŸ©é˜µ A:\n", A)
print("çŸ©é˜µ B:\n", B)

# 1. é€å…ƒç´ ä¹˜æ³• (*)
# å¯¹åº”ä½ç½®ç›¸ä¹˜ï¼š1*5, 2*6, ...
print("\né€å…ƒç´ ä¹˜æ³• (A * B):\n", A * B)

# 2. çŸ©é˜µä¹˜æ³• (@)
# è¡Œä¹˜åˆ—ï¼š1*5+2*7, 1*6+2*8, ...
print("\nçŸ©é˜µä¹˜æ³• (A @ B):\n", A @ B)

# 3. å‘é‡çš„ç‚¹ç§¯
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
# 1*4 + 2*5 + 3*6 = 32
print(f"\nå‘é‡å†…ç§¯ (v1 @ v2): {v1 @ v2}")
```

## numpy.linalg å·¥å…·ç®±

|å‡½æ•°|	æè¿°|	æ•°å­¦å«ä¹‰|
|---|---|---|
|np.linalg.inv(A)|	æ±‚é€†çŸ©é˜µ	|Aâˆ’1A^{-1} A âˆ’ 1|
|np.linalg.det(A)|	æ±‚è¡Œåˆ—å¼	|$|
|np.linalg.norm(A)|	æ±‚èŒƒæ•°ï¼ˆæ¨¡é•¿ï¼‰|	âˆ£Aâˆ£|A| âˆ£ A âˆ£|
|np.linalg.eig(A)|	ç‰¹å¾å€¼åˆ†è§£|	Av=Î»vAv = \lambda v A v = Î» v|
|np.linalg.svd(A)|	å¥‡å¼‚å€¼åˆ†è§£|	SVD|
|np.linalg.solve(A, b)|	è§£æ–¹ç¨‹ç»„|	Ax=bAx = b A x = b|


```python
# ç³»æ•°çŸ©é˜µ A
A = np.array([[3, 1],
              [1, 2]])
# ç»“æœå‘é‡ b
b = np.array([9, 8])

# é”™è¯¯åšæ³•ï¼ˆä¸æ¨èï¼‰ï¼šæ˜¾å¼æ±‚é€†
# x_bad = np.linalg.inv(A) @ b

# æ­£ç¡®åšæ³•ï¼šä½¿ç”¨ solve
x = np.linalg.solve(A, b)

print(f"è§£å¾—: x={x[0]:.1f}, y={x[1]:.1f}")

# éªŒè¯ç»“æœ: A @ x åº”è¯¥ç­‰äº b
# ä½¿ç”¨ np.allclose æ¯”è¾ƒæµ®ç‚¹æ•°ï¼Œä¸è¦ç”¨ ==
is_correct = np.allclose(A @ x, b)
print(f"éªŒè¯ç»“æœæ˜¯å¦æ­£ç¡®: {is_correct}")
```

```python
# åˆ›å»ºä¸€ä¸ªå¯¹ç§°çŸ©é˜µï¼ˆå¯¹ç§°çŸ©é˜µçš„ç‰¹å¾å‘é‡æ€»æ˜¯æ­£äº¤çš„ï¼Œæ€§è´¨å¾ˆå¥½ï¼‰
M = np.array([[1, 2],
              [2, 1]])

# è®¡ç®—ç‰¹å¾å€¼å’Œç‰¹å¾å‘é‡
eigenvalues, eigenvectors = np.linalg.eig(M)

print("ç‰¹å¾å€¼:", eigenvalues)
print("ç‰¹å¾å‘é‡ (æ¯ä¸€åˆ—æ˜¯ä¸€ä¸ªå‘é‡):\n", eigenvectors)

# éªŒè¯: M @ v = lambda * v
v0 = eigenvectors[:, 0]
lambda0 = eigenvalues[0]

lhs = M @ v0           # å·¦è¾¹: çŸ©é˜µ x å‘é‡
rhs = lambda0 * v0     # å³è¾¹: æ ‡é‡ x å‘é‡

print("\néªŒè¯ç‰¹å¾å€¼æ€§è´¨:", np.allclose(lhs, rhs))
```

## é«˜ç»´çŸ©é˜µä¹˜æ³•

```python
# æ¨¡æ‹Ÿ 10 ä¸ªæ ·æœ¬ï¼Œæ¯ä¸ªæ ·æœ¬æ˜¯ä¸€ä¸ª 2x3 çš„çŸ©é˜µ
A_batch = np.random.rand(10, 2, 3)

# æ¨¡æ‹Ÿ 10 ä¸ªæ ·æœ¬ï¼Œæ¯ä¸ªæ ·æœ¬æ˜¯ä¸€ä¸ª 3x4 çš„çŸ©é˜µ
B_batch = np.random.rand(10, 3, 4)

# æˆ‘ä»¬æƒ³å¯¹è¿™ 10 ä¸ªæ ·æœ¬åˆ†åˆ«åšçŸ©é˜µä¹˜æ³•
# ç»“æœåº”è¯¥æ˜¯ (10, 2, 4)
C_batch = A_batch @ B_batch

print(f"A å½¢çŠ¶: {A_batch.shape}")
print(f"B å½¢çŠ¶: {B_batch.shape}")
print(f"ç»“æœå½¢çŠ¶: {C_batch.shape}")

# éªŒè¯ç¬¬ä¸€ä¸ªæ ·æœ¬
# C_batch[0] åº”è¯¥ç­‰äº A_batch[0] @ B_batch[0]
print("éªŒè¯ Batch 0:", np.allclose(C_batch[0], A_batch[0] @ B_batch[0]))
```
